```python
# Import the libraries
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import confusion_matrix
import itertools
```

```python
# Create function for confusion matrix
def plot_confusion_matrix(cm, classes, normalize=False,title='Confusion Matrix', cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normazilation can be apploed by setting 'normalize=True'
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print('Normalized confusion matrix')
    else:
        print('Confusion matrix, without Normalization')
    
    print(cm)
    
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.0
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment='center',
                 color='white' if cm[i, j] > thresh else 'black')
    
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```

```python
# Load in the data
data = tf.keras.datasets.fashion_mnist
(X_train_raw, y_train), (X_test_raw, y_test) = data.load_data()

# Scale the data to 0-1
X_train = X_train_raw / 255.0
X_test = X_test_raw / 255.0

# Havce a look at the shape
print('Train data shape:', X_train.shape)
print('Test data shape:', X_test.shape)
```

```python
# Have a look at a random image
i = np.random.choice(X_test.shape[0])
print(f'Index: {i}')
plt.imshow(X_test[i], cmap='gray');
```

### Try solve fashion MNIST with ANN

```python
# Define the layers
input_layer = tf.keras.layers.Flatten(input_shape=(28,28)) # Converts image data to 'flat' vector representation
hidden_one = tf.keras.layers.Dense(100, activation='relu') # The first hidden layer with 128 neurons
dropout = tf.keras.layers.Dropout(0.2) # Randomly sets 20% of the neurons to 0 during each forward pass
output_layer = tf.keras.layers.Dense(10, activation='softmax')
```

```python
# Create the model
model = tf.keras.models.Sequential([input_layer, 
                                    hidden_one, 
                                    dropout, 
                                    output_layer])
```

```python
# Define how the model should be trained
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Create a learning rate scheduler
def schedule(epoch, lr):
    print('Learning Rate:', lr)
    return lr * 0.9
    
# Use the scheduler
scheduler = tf.keras.callbacks.LearningRateScheduler(schedule)
```

```python
# Train the model
result = model.fit(X_train, y_train,
                   validation_data=(X_test, y_test),
                   epochs=15, callbacks=[scheduler])
```

```python
# Print the results
print("Train score:", model.evaluate(x=X_train, y=y_train, verbose=0)[1]);
print("Test score:", model.evaluate(x=X_test, y=y_test, verbose=0)[1]);
```

```python
# Plot the cost function at each itteration
# Note: model.fit() returns an object that contains the history of the training process
plt.plot(result.history['loss'], label='loss');
plt.plot(result.history['val_loss'], label='val_loss');
plt.legend();
```

```python
# Plot the cost function at each itteration
# Note: model.fit() returns an object that contains the history of the training process
plt.plot(result.history['accuracy'], label='loss');
plt.plot(result.history['val_accuracy'], label='val_loss');
plt.legend();
```

```python
# Plot confusion matrix
p_test = model.predict(X_test).argmax(axis=1)
cm = confusion_matrix(y_test, p_test)
plot_confusion_matrix(cm, list(range(10)))
```

### Try solve fashion MNIST with CNN

```python
# Import libraries
from tensorflow.keras.layers import Input, Conv2D, Dense, Flatten, Dropout
from tensorflow.keras.models import Model
```

```python
# Pre-process data to add a new dimension (CNN expects H x W x C)
X_train_processed = np.expand_dims(X_train, -1)
X_test_processed = np.expand_dims(X_test, -1)

# Confirm that it has 3 dimensions
print(X_train_processed.shape)
print(X_test_processed.shape)
```

```python
# Get the number of classes
K = len(set(y_train))
print('Number of classes:', K)
```

```python
# Define the layers
i = Input(shape=X_train_processed[0].shape)
# 32 feature maps, 3x3 filter dimension, activation=relu, mode=valid (default)
# Using strides = 2 means the size of the image will be halved after every convolution
x = Conv2D(32, (3,3), strides=2, activation='relu')(i)
x = Conv2D(64, (3,3), strides=2, activation='relu')(x)
x = Conv2D(128, (3,3), strides=2, activation='relu')(x)
x = Flatten()(x)
x = Dropout(0.2)(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.2)(x)
x = Dense(K, activation='softmax')(x)
```

```python
# Create the model
model = Model(i, x)
```

```python
# Define how the model should be trained
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Create a learning rate scheduler
def schedule(epoch, lr):
    print('Learning Rate:', lr)
    return lr * 0.9
    
# Use the scheduler
scheduler = tf.keras.callbacks.LearningRateScheduler(schedule)
```

```python
# Train the model
result = model.fit(X_train_processed, y_train,
                   validation_data=(X_test_processed, y_test),
                   epochs=15, callbacks=[scheduler])
```

```python
# Print the results
print("Train score:", model.evaluate(x=X_train_processed, y=y_train, verbose=0)[1]);
print("Test score:", model.evaluate(x=X_test_processed, y=y_test, verbose=0)[1]);
```

```python
# Plot the cost function at each itteration
# Note: model.fit() returns an object that contains the history of the training process
plt.plot(result.history['loss'], label='loss');
plt.plot(result.history['val_loss'], label='val_loss');
plt.legend();
```

```python
# Plot the cost function at each itteration
# Note: model.fit() returns an object that contains the history of the training process
plt.plot(result.history['accuracy'], label='loss');
plt.plot(result.history['val_accuracy'], label='val_loss');
plt.legend();
```

```python
# Plot confusion matrix
p_test = model.predict(X_test_processed).argmax(axis=1)
cm = confusion_matrix(y_test, p_test)
plot_confusion_matrix(cm, list(range(10)))
```

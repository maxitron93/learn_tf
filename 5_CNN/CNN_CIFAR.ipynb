```python
# Import libraries
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import itertools
from tensorflow.keras.layers import Input, Conv2D, Dense, Flatten, Dropout, GlobalMaxPooling2D
from tensorflow.keras.models import Model
```

```python
# Create function for confusion matrix
def plot_confusion_matrix(cm, classes, normalize=False,title='Confusion Matrix', cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normazilation can be apploed by setting 'normalize=True'
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print('Normalized confusion matrix')
    else:
        print('Confusion matrix, without Normalization')
    
    print(cm)
    
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)
    
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.0
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment='center',
                 color='white' if cm[i, j] > thresh else 'black')
    
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```

```python
# Load in the data
data = tf.keras.datasets.cifar10
(X_train_raw, y_train), (X_test_raw, y_test) = data.load_data()
```

```python
# Scale the data to 0-1
X_train = X_train_raw / 255.0
X_test = X_test_raw / 255.0
```

```python
# Flatten the labels (just to get it in the right format)
y_train, y_test = y_train.flatten(), y_test.flatten()
```

```python
print(X_train.shape)
```

```python
print(y_train.shape)
```

```python
# Print the number of classes
K = len(set(y_train))
print('Number of classes:', K)
```

```python
# Define the layers
i = Input(shape=X_train[0].shape)
# 32 feature maps, 3x3 filter dimension, activation=relu, mode=valid (default)
# Using strides = 2 means the size of the image will be halved after every convolution
x = Conv2D(32, (3,3), strides=2, activation='relu')(i)
x = Conv2D(64, (3,3), strides=2, activation='relu')(x)
x = Conv2D(128, (3,3), strides=2, activation='relu')(x)
x = Flatten()(x)
x = Dropout(0.5)(x)
x = Dense(1024, activation='relu')(x)
x = Dropout(0.2)(x)
x = Dense(K, activation='softmax')(x)
```

```python
# Create the model
model = Model(i, x)
```

```python
# Define how the model should be trained
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Create a learning rate scheduler
def schedule(epoch, lr):
    print('Learning Rate:', lr)
    return lr * 0.9
    
# Use the scheduler
scheduler = tf.keras.callbacks.LearningRateScheduler(schedule)
```

```python
# Train the model
result = model.fit(X_train, y_train,
                   validation_data=(X_test, y_test),
                   epochs=15, callbacks=[scheduler])
```

```python
# Print the results
print("Train score:", model.evaluate(x=X_train, y=y_train, verbose=0)[1]);
print("Test score:", model.evaluate(x=X_test, y=y_test, verbose=0)[1]);
```

```python
# Plot the cost function at each itteration
# Note: model.fit() returns an object that contains the history of the training process
plt.plot(result.history['loss'], label='loss');
plt.plot(result.history['val_loss'], label='val_loss');
plt.legend();
```

```python
# Plot the cost function at each itteration
# Note: model.fit() returns an object that contains the history of the training process
plt.plot(result.history['accuracy'], label='loss');
plt.plot(result.history['val_accuracy'], label='val_loss');
plt.legend();
```

```python
# Plot confusion matrix
p_test = model.predict(X_test).argmax(axis=1)
cm = confusion_matrix(y_test, p_test)
plot_confusion_matrix(cm, list(range(K)))
```

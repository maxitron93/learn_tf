```python
# Import libraries
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input, SimpleRNN, GRU, LSTM, Dense, Flatten, GlobalMaxPool1D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import SGD, Adam
```

### Make the synthetic dataset

```python
# Define the dimensions
T = 20
D = 1
X_raw = []
Y_short = [] # short distance labels
Y_long = [] # long distance labels
```

```python
# Create a function to return a label
def get_label(x, i1, i2, i3):
    # x = sequence
    if x[i1] < 0 and x[i2] < 0 and x[i3] < 0:
        return 1
    if x[i1] < 0 and x[i2] > 0 and x[i3] > 0:
        return 1
    if x[i1] > 0 and x[i2] < 0 and x[i3] > 0:
        return 1
    if x[i1] > 0 and x[i2] > 0 and x[i3] < 0:
        return 1
    return 0
```

```python
# Create the data
for t in range(5000):
    x = np.random.randn(T)
    X_raw.append(x)
    y_short = get_label(x, -1, -2, -3) # Create the label based on the last three datapoints (short-distance)
    y_long = get_label(x, 0, 1, 2) # Create the label based on the first three datapoints (long-distance)
    Y_short.append(y_short)
    Y_long.append(y_long)
```

```python
# Change to correct format
X = np.array(X_raw)
y_short = np.array(Y_short)
y_long = np.array(Y_long)
N = len(X)
```

### Try a linear model
We know this won't work because it's not linearly seperable

```python
# Define the layers of the model
i = Input(shape=(T,))
x = Dense(1, activation='sigmoid')(i)

# Create the model
model = Model(i, x)

# Define how the model should be trained
model.compile(optimizer=Adam(lr=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the mdoel
result = model.fit(x=X, y=y_short,
                   epochs=100,
                   validation_split=0.25)
```

```python
# Plot the cost function at each itteration
plt.plot(result.history['accuracy'], label='train_accuracy');
plt.plot(result.history['val_accuracy'], label='val_accuracy');
plt.legend();
```

### Try an ANN

```python
# Define the layers of the model
i = Input(shape=(T,))
x = Dense(50, activation='relu')(i)
x = Dense(1, activation='sigmoid')(x)

# Create the model
model = Model(i, x)

# Define how the model should be trained
model.compile(optimizer=Adam(lr=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the mdoel
result = model.fit(x=X, y=y_short,
                   epochs=100,
                   validation_split=0.25)
```

```python
# Plot the cost function at each itteration
plt.plot(result.history['accuracy'], label='train_accuracy');
plt.plot(result.history['val_accuracy'], label='val_accuracy');
plt.legend();
```

This works okay, but the accuracy isn't as high as it could be because the ANN can't take sequence data into account

### Try a SimpleRNN

#### Dataset where the pattern is at the end (only short-distance 'memory' required)

```python
# Reshape the data
inputs = np.expand_dims(X, -1)

# Define the layers of the model
i = Input(shape=(T,D))
x = SimpleRNN(5, activation='tanh')(i)
x = Dense(1, activation='sigmoid')(x)

# Create the model
model = Model(i, x)

# Define how the model should be trained
model.compile(optimizer=Adam(lr=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the mdoel
result = model.fit(x=inputs, y=y_short,
                   epochs=100,
                   validation_split=0.25)
```

```python
# Plot the cost function at each itteration
plt.plot(result.history['accuracy'], label='train_accuracy');
plt.plot(result.history['val_accuracy'], label='val_accuracy');
plt.legend();
```

The SimpleRNN works well because the pattern is at the end, so long-term 'memory' isn't required, so the vanishing gradient problem isn't an issue.

#### Dataset where the pattern is at the start (long-distance 'memory' required)

```python
# Define the layers of the model
i = Input(shape=(T,D))
x = SimpleRNN(5, activation='tanh')(i)
x = Dense(1, activation='sigmoid')(x)

# Create the model
model = Model(i, x)

# Define how the model should be trained
model.compile(optimizer=Adam(lr=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the mdoel
result = model.fit(x=inputs, y=y_long,
                   epochs=100,
                   validation_split=0.25)
```

```python
# Plot the cost function at each itteration
plt.plot(result.history['accuracy'], label='train_accuracy');
plt.plot(result.history['val_accuracy'], label='val_accuracy');
plt.legend();
```

Now that the pattern is at the start, the SimpleRNN doesn't work well because long-term 'memory' is required - and SimpleRNN isn't good at that because of the vanishing gradient problem.

### Try a LSTM

#### Dataset where the pattern is at the end (only short-distance 'memory' required)

```python
# Define the layers of the model
i = Input(shape=(T,D))
x = LSTM(5, activation='tanh')(i)
x = Dense(1, activation='sigmoid')(x)

# Create the model
model = Model(i, x)

# Define how the model should be trained
model.compile(optimizer=Adam(lr=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the mdoel
result = model.fit(x=inputs, y=y_short,
                   epochs=100,
                   validation_split=0.25)
```

```python
# Plot the cost function at each itteration
plt.plot(result.history['accuracy'], label='train_accuracy');
plt.plot(result.history['val_accuracy'], label='val_accuracy');
plt.legend();
```

#### Dataset where the pattern is at the start (long-distance 'memory' required)

```python
# Define the layers of the model
i = Input(shape=(T,D))
x = LSTM(5, activation='tanh')(i)
x = Dense(1, activation='sigmoid')(x)

# Create the model
model = Model(i, x)

# Define how the model should be trained
model.compile(optimizer=Adam(lr=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the mdoel
result = model.fit(x=inputs, y=y_long,
                   epochs=100,
                   validation_split=0.25)
```

```python
# Plot the cost function at each itteration
plt.plot(result.history['accuracy'], label='train_accuracy');
plt.plot(result.history['val_accuracy'], label='val_accuracy');
plt.legend();
```

This still works okay. It's able to 'remember' long-distance patterns. This is because LSTM is less effected by the vanishing gradient problem relative to the SimpleRNN. However, the LSTM still has its limits. If the pattern is far-away enough, it will not be able to remember it.

### Try LSTM using all hidden states

```python
# Define the layers of the model
i = Input(shape=(T,D))
x = LSTM(5, activation='tanh', return_sequences=True)(i)
x = GlobalMaxPool1D()(x)
x = Dense(1, activation='sigmoid')(x)

# Create the model
model = Model(i, x)

# Define how the model should be trained
model.compile(optimizer=Adam(lr=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Train the mdoel
result = model.fit(x=inputs, y=y_long,
                   epochs=100,
                   validation_split=0.25)
```

```python
# Plot the cost function at each itteration
plt.plot(result.history['accuracy'], label='train_accuracy');
plt.plot(result.history['val_accuracy'], label='val_accuracy');
plt.legend();
```

We can see this performed better than the LSTM without using all the hidden states.
